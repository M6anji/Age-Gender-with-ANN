```python
import cv2
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers , models
import numpy as np
import pandas as pd
```

```python
face_detector = cv2.CascadeClassifier (cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
Gender_model = models.load_model("Gender.h5")
Age_model = models.load_model("Age.h5")


vid = cv2.VideoCapture(0)
while (True):
    ret , frame = vid.read()
    Gray_color = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    face = face_detector.detectMultiScale(Gray_color)
    for (x , y , w , h) in face:
        cv2.rectangle(frame , (x , y ) , (x + w , y +h) , (255 , 0 , 0) , 3)
        Face = frame[y:y+h , x:x+w]
        face_resized = cv2.resize(Face, (64, 64),interpolation =cv2.INTER_AREA)
        face_flatten = face_resized / 255.0
        face_flatten = face_flatten.flatten()
        face_flatten = np.expand_dims(face_flatten, axis=0)
        
        pred = Gender_model.predict(face_flatten) 
        print (pred)
        if pred[0][0] > 0.5:
            label = 'Man'
        else:
            label = 'Woman'
        print (label)
        cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0 , 255 , 0) , 2)
        
        output = Age_model.predict(face_flatten)
        print(output)
        max_output = np.argmax(output)
        print(max_output)
        category_name = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-More']
        print(category_name[max_output])    
        
        cv2.putText(frame, category_name[max_output], (x, y - 30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0 , 255 , 0) , 2)

    cv2.imshow("Mehdi_Ganji_Project" , frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
vid.release()
cv2.destroyAllWindows()
```
